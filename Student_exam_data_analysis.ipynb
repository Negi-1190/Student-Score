```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/Expanded_data_with_more_features.csv.zip')
df.info()
```

```python
#drop unname column
df=df.drop('Unnamed: 0',axis=1)
```

```python
df.head()
```

```python
#change weekly study hrs column
df['WklyStudyHours']=df['WklyStudyHours'].str.replace('5 - 10	','05-10')
```

```python
df.head()
```

```python
# Gender distrbution by ploting
ax=sns.countplot(data=df,x='Gender')
plt.title('Gender distrbution')
plt.figure(figsize=(4,4))
ax.bar_label(ax.containers[0])

```

**from the above chart we analys that the number of female is more than the number of male\\**

```python
xs=df.groupby('ParentEduc').agg({'MathScore':'mean', 'ReadingScore':'mean','WritingScore':'mean'})
xs
plt.title('Relationship between parents education & student')

sns.heatmap(xs,annot=True)
plt.figure(figsize=(4,4))
plt.show()
```

from the above analysis we saw that parents education is impacting their scores

```python
xs1=df.groupby('ParentMaritalStatus').agg({'MathScore':'mean', 'ReadingScore':'mean','WritingScore':'mean'})
plt.title('Relationship between parents Marital status & student Score')

sns.heatmap(xs1,annot=True)
plt.figure(figsize=(4,4))
plt.show()
```

form the above chart we analysis that parents marrital status is not making any impact on student scores
